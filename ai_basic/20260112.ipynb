```python
# 지도학습
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# 데이터 준비 : train data, test data 
# 특성 데이터, 타깃 데이터

# 특성
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # 학습 데이터 2차원 형식으로 형태 변환
y = np.array([50, 55, 65, 70, 75]) # 라벨(타깃)
```

```python
# 학습 모델 생성 (학습모델 객체 생성) -> 학슴

model = LinearRegression()

# 훈련 : fit(특성, 타깃)

model.fit(X,y)
```

```python
# 학습 결과 지표
# 기울기

m = model.coef_[0] # 기울기
b = model.intercept_ # 절편

print('기울기:', m, '절편:', b)
```

```python
# 학습된 모델로 값을 예측: rpedict
y_pred = model.predict(X)
print(y_pred)


```

```python
# 성능 지표 
mse = mean_squared_error(X, y_pred)
mae = mean_absolute_error(X, y_pred)
r2 = r2_score(X, y_pred)
rmse = np.sqrt(mse)

print('mse:', mse)
print('mae:', mae)
print('r2:', r2)
print('rmse:', rmse)
```

```python
# 차트: scatter, plot

plt.scatter(X,y)
plt.plot(X,y_pred, color = 'red')
```

```python
# 예측 : 6, 8
new_hour = np.array([6,8]).reshape(-1, 1)
predict_score = model.predict(new_hour)
print(predict_score)
```
