```python
from bs4 import BeautifulSoup
```

```python
# 문자열
html_doc = """
<html>
    <head>
        <title>Example Page</title>
    </head>
    <body>
        <h1 class = "description">Welcome to Web Scraping</h1>
        <p class="description big">This is an example paragraph.</p>
        <a href="https://example.com1">Visit Example1</a>
        <a href="https://example.com2">Visit Example2</a>
        <a href="https://example.com3">Visit Example3</a>
        <a href="https://example.com4">Visit Example4</a>
    </body>
</html>
"""
```

```python
soup = BeautifulSoup(html_doc, "html.parser")
```

```python
print(soup.title.string)
print(soup.h1.string)
print(soup.a['href'])
print(soup.p)
print(soup.p['class'])
```

```python
# find(tag), find_all(tag)

a_tag = soup.find('a')
print(a_tag)

a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []
for a_tag in a_tags:
    print(a_tag['href'])
    links.append(a_tag['href'])

print(links)
```

```python
desc_tag = soup.select_one('.description') # #=>id, .=>class, *=>all
print(desc_tag)

desc_tags = soup.select('.description')
print(desc_tags)

a_tag2 = soup.select('a')
print(a_tag2)
```

### SQLite: 관계형 데이터베이스

```python
import sqlite3
```

```python
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공')
conn.close()
```

```python
# 관계형 DB => 데이블 구조로 저장, 관리
# 저장할 수 있는 공간 생성 => 테이블 생성

sql_create = ''' 
create table user (
    id integer primary key autoincrement,
    name text not null,
    age integer,
    city text
)
'''

# 테이블 생성
# 1. 데이터베이스 연결
conn = sqlite3.connect('first.db')

# 2. sql실행 시키기 위한 객체 필요
cursor = conn.cursor()

# 3. sql 실행
cursor.execute(sql_create)
print('테이블 생성!')

# 4. 연결 종료
conn.close()
```

```python
# 데이터 입력 : 행 단위 입력 = >(이름 , 나이, 거주지)

sql_insert = '''
INSERT INTO user (name, age, city)
VALUES (?, ?, ?)
'''

conn = sqlite3.connect('first.db')
cursor = conn.cursor()
#cursor.execute(sql_insert, ('cool', 20, 'Seoul'))

# 입력, 수정, 삭제 : DML => 작업 완료 또는 롤백을 지정해주어야 한다.
# commit : 작업의 완료를 알려주는 메서드

data = [
    ('cool1', 20, 'Seoul'),
    ('cool2', 30, 'Seoul'),
    ('cool3', 40, 'Seoul')
]

cursor.executemany(sql_insert, data)

conn.commit() # 작업 끝났으니 파일에 저장해!!
conn.close()
```

```python
# 검색 (질의) : 데이터의 검색 => 행단위 검색

sql_select = 'select * from user'
sql_select_where = 'select * from user where id > 2'
sql_select_new = 'select name, city from user where age < 30'


conn = sqlite3.connect('first.db')
cursor = conn.cursor()
cursor.execute(sql_select)

# sql의 결과 => fetchall(), fetchone()
rows = cursor.fetchall()
print(rows)

for row in rows :
    print(row)

# for id, age, name, city in rows :
#     print(id, age, name, city)

conn.close()
```

```python
# 데이터 수정
# 행을 찾고 -> 행의 컬럼값을 변경

sql_update = '''
    update user set age = ? where id = ?
'''

conn = sqlite3.connect('first.db')
cursor = conn.cursor()
cursor.execute(sql_update, (25, 4))
conn.commit()
print('update')

conn.close()

```

```python
# 데이터 삭제 : 행단위 삭제

sql_delete = 'delete from user where id =?'

conn = sqlite3.connect('first.db')
cursor = conn.cursor()
cursor.execute(sql_delete, (4,))
conn.commit()

# commit 후
print('삭제완')

conn.close()
```
