```python
import pandas as pd
```

```python
# 결측치 처리

# 데이터프레임 생성
data = {'Name': ['Alice', 'Bob', None],
'Age': [25, None, 30],
'Score': [85, 90, None]}

df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.isnull()
df
```

```python
# 결측치가 있는 행을 제거

df_missing_rows = df.dropna()
print(df_missing_rows)
print('-'*30)
df_missing_cols = df.dropna(axis = 1)
print(df_missing_cols)
```

```python
# 결측치에 값을 대체
# Name: 최빈값
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])
# Age:평균값
df['Age'] = df['Age'].fillna(df['Age'].mean())
# Score: 중앙값
df['Score'] = df['Score'].fillna(df['Score'].median())


df
```

```python
# 이상치 처리

data = {
    'score' : [70, 75, 80, 85, 90, 150]
}

df = pd.DataFrame(data)
df
```

```python
#IQR

Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)

IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

```python
# 이상치의 범위

low_ = Q1 - 1.5*IQR
up_ = Q3 + 1.5*IQR

print(f'이상치의 범위 : {low_}보다 작은 값, {up_}값조다 큰 값')
```

```python
# 이상치 찾기

df_1 = df[(df['score']<low_) | (df['score']>up_)]
df_1 
```

```python
df_2 = df[(df['score'] >= low_) & (df['score'] <= up_)]
df_2 
```

```python
# 이상치 데이터 변경
df['score'] = df['score'].astype(float)
df.loc[df['score']<low_, 'score'] = low_
df.loc[df['score']>up_, 'score'] = up_

df
```

```python
# merge : 열단위 데이터 병합 => 옆으로 붙인다!!!

import pandas as pd

# 고객 정보 데이터프레임
customers = pd.DataFrame({
'customer_id': [1, 2, 3],
'name': ['Alice', 'Bob', 'Charlie']
})
# 주문 정보 데이터프레임
orders = pd.DataFrame({
'order_id': [101, 102, 103],
'customer_id': [1, 2, 4],
'product': ['Laptop', 'Tablet', 'Smartphone']
})

```

```python
# 공통된 customer_id 데이터를 병합 : Inner.Join 
result1 = pd.merge(customers, orders, on = 'customer_id', how = 'inner') # 순서 중요
print(result1)
```

```python
# 고객별 주문 내역 : 구매 이력이 없는 고객도 표현

result2 = pd.merge(customers, orders, on = 'customer_id', how = 'left' )
print(result2)
```
