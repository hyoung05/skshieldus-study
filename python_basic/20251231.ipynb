반복문: while

```python
# 반복문 : while conditions(True/False) :
#          반복구문   
#          ...

cnt = 0

while cnt < 5 :
    print(cnt)
    cnt = cnt + 1 # 증가식 
```

```python
# 1~0 까지 출력, 홀수만 출력
cnt = 0

while cnt < 10 :
    cnt = cnt + 1 # 증가식
    if cnt % 2 == 1 :
        print(cnt) 
```

```python
# continue : 반복문 내에서 만나면 아래 코드를 실행하지 않고 블록 종료 => 조건식으로 간다

cnt = 0

while cnt < 10 :
    cnt += 1 # 증가식

    if cnt % 2 == 0 :
        continue
    print(cnt) 
```

```python
# break : 반복문 내에서 이 키워드를 만나면 반복문을 종료시킨다

cnt = 0

while True :
    cnt += 1
    print(cnt)

    if cnt > 5 :
        break
```

```python
# 5명의 학생의 영어시험 => 60 점 이상 pass 출력
# 5명의 학생의 영어 점수의 총점과 평균

scores = [80, 50, 70, 100, 55]
num = 0
sum = 0

for s in scores :
    num = num + 1
    # 합을 더해나가는 식
    sum = sum + s
    
    if s >= 60 :
        print(num, s, 'pass')
    else :
        print(num, s, 'Fail')
    
print('총점:', sum)
print('평균:', sum/len(scores))
```

```python
# range 함수 : 일련번호를 담은 리스트 생성해주는 함수
# range(start, end, step)
# range(end)

for num in range(10) : # 0 ~ 10-1
    print(num, end = '   ')
```

```python
#구구단: 2~9단 => 2~9
#각 단: 1~9까지 X => 1~9

for dan in range(2, 10) :
    print(dan, '단')
    # 이 블럭 안에서 각 단의 숫자에 1~9 까지 곱하는 연산을 해주어야 한다
    for i in range (1, 10) :
        print(f'{dan}X{i} = {dan*i}', end= ' ')
```

```python
# 함수 정의 : 두개 숫자를 받아서 더하기한 결과를 반환하는 함수
# 이름: add
# 매개변수 : n1, n2
# 처리: n1 + n2
# 반환: return n1 + n2

def add(n1, n2) :
    result = n1 + n2
    return result

add(10,20)
```

```python
# 매개변수사 몇개인지 몰라???!!
#전달되는 모든 매개변수 숫자를 더합 닶을 반환하는 함수

def add_all_sum(*args) :
    print(args, type(args))

    sum = 0

    for num in args:
        sum += num
    return sum


add_all_sum(1,2,3,4)
```

```python
# 연산기호와 전달되는 숫자를 받아서
# 연산의 결과를 반환하는 함수

def proc_type(type, *args) : 
    print(type, args)
    if type == '+':
        result = 0
        for num in args:
            result += num
        return result

    elif type == '*':
        result = 1
        for num in args :
            result *=num
        return result

    else :
        print('사용 가능한 연산이 아닙니다. 다시 입력해주세요.')
```

```python
proc_type('+', 1,2,3)
```

```python
# filter : 원본 리스트의 데이터들을 필터링해서 새로운 리스트를 만들어서 반환하는 함수
# 필터링 하는 함수를 매개 변수로 받는다!
# filter(function, list)
# 판별 함수

def positive_fn(x) :
    return x>0
```

```python
list(filter(positive_fn, [3,1,7,-8,0,-1, -3, 10]))
```

```python
list(filter(lambda x : x>0, [3,1,7,-8,0,-1, -3, 10]))
```

```python
# map : 원본 리시트의 값을 기반으로 연상한 결과 값들을 이용해서 새로운 리스트를 생성하고 반환

list (map (lambda x : x*10, [1,2,3,4,5]))
```
