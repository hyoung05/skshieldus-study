```python
#if : 분기 => ~라면, ~가 아니라면
#성인 판단 => 성인이라면 성인이 아니라면

age = 25

# 성이라면 -> 환영 메시지, 아니면 -> 킥~
if age >= 18:
    print('환영합니다.')
    print('서버에 접속합니다. 게입을 시작합니다!')
else :
    print('킥 ~~!!! 오면 안됩니다!!')
```

```python
# 짝수 / 홀수 판단 => 숫자 % 2 : 1 -> 홀수, 0 -> 짝수
# 숫자 데이터 기준 
# 짝수 홀수 비교 
# => 짝수/홀수 출력

num = 9
if num % 2 == 0 :
    print("짝수")
else:
    print("홀수")
```

```python
# 양수, 음수, 0 판단해서 결과 출겨
# 숫자 데이터
# 양수 : num > 0
# 음수 : num < 0
# 0 : num > 0 아니고 num < 0

num1 = 0

if num1> 0 :
    print('양수')
elif num1< 0 :
    print('음수')
else:
    print('0')
        
```

```python
# 운전명허 시험

score = 69
if score >= 70 :
    print('PASS')
else : 
    print('FAIL')

check_msg = 'PASS'

if score < 70:
    check_msg = 'FAIL'

print(check_msg)
```
