01 숫자를 입력 받아 짝수, 홀수를 판별하는 프로그램

```python
num = int(input(print('숫자를 입력해주세요')))
print(f':{num}')

if num % 2 == 0:
    print('짝수')
else :
    print('홀수')
```

02 숫자를 입력 받아 양수, 음수, 0를 판별하는 프로그램

```python
num1 = int(input(print('숫자를 입력해주세요')))
print(f':{num1}')

if num1 > 0:
    print('양수')
elif num1 < 0:
    print('음수')
else:
    print('0')
```

03 독감예방 접종이 가능한지 여부를 확인하는 프로그램

```python
num2 = int(input(print('나이를 입력하세요.')))
print(f'{num2}는')

if num2 < 15 | num2 >= 65 :
    print('무료예방접종이 가능합니다.')
else:
    print('무료예방접종 대상이 아닙니다.')
```

04 특정 연도 건간검진 대상 여부 판별 및 검진 종류 확인

```python
num3 = int(input(print('몇 년생입니까?')))
print(f':{num3}')
num4 = int(input(print('올해는 몇 년도 입니까?')))
print(f':{num4}')

age = (num4-num3)+1

print('-'*30)
print(f'{age}세는')

if age >= 20:
    if num3 % 2 == 0 & num4 % 2 == 0:
        print('짝수 해 건강검진 대상입니다.')
        
        if age >= 40 :
            print('암 검사 무료 대상입니다.')

        else :
            print('')

    elif num3 % 2 == 1 &  num4 % 2 == 1:
        print('홀수 해 건강검진 대상입니다.')
        
        if age >= 40 :
            print('암 검사 무료 대상입니다.')

        else:
            print('')
else:
    print('건강검진 대상이 아닙니다')
```
