### 클래스와 객체지향

```python
# 객체 지향 프로그래밍 : 객의 개념을 도입해서 사용
# 객체: 변수들, 함수들 포함하는 메모리 묶음
# class =>변수들과 함수들이 어게 구성되는지 정의하는 설계
# class에서 정의하는 것 -> 변수들 함수들
# class => 새로운 타입을 정의하는 것
# 객체 생성 => 변수 = 클래스이름()


#객체 생성을 위한 class 정의 : 객체가 가지는 변수/함수들의 규칙 정의
# class 타입 이름():
#    변수들...
#    함수들...

class Cookie():
    pass
```

```python
# 사칙 연산을 하는 클래스: 클래스 이름, 변수, 함수 초기화 함수
# FourCal => 클래스의 이름은 대문자로 작성
# 변수: 두개의 정수
# 함수 : 사칙연산 + - * /

class FourCal:
    # 클래스의 멤버: 변수, 함수

    # 변수의 생성과 초기화
    # 초기화(변수 생성-> 최초 데이터 할당) =>생성자 함수, 생성자
    def __init__(self, first, second):
        self.first = first
        self.second = second

    # 변수의 초기화(변수 선언->최초 데이터의 할당)
    # 클래스에서 함수를 정의 할 때는 self 매개변수 반드시 받는다!
    # self => 객체의 자기 자신을 가르키는 변수(객체 자기 자신의 메모리 주소)

    def setData(self, first, second) :
        self.first = first
        self.second = second

        
        # 메서드 : 객체 안의 멤버인 함수의 명칭
        # 사칙연산을 하는 메서드 정의

    def add(self):
        return self.first + self.second

    def sub(self) :
        return self.first - self.second

    def mul(self) :
        return self.first * self.second

    def div(self) :
        return self.first / self.second
        
    def print_result(self):
        print('사칙연산의 결과')
        print('='*30)
        print('first:', self.first, 'second:', self.second)
        print(f'덧셈 :{self.add()}' )
        print(f'뺄셈 :{self.sub()}' )
        print(f'곱셈 :{self.mul()}' )
        print(f'나누기 :{self.div()}' )
```

```python
fc1 = FourCal(100, 30)
fc2 = FourCal(1000, 30)
```

```python
fc1.print_result()
```

```python
print(type(fc1), fc1.first, fc1.second, fc1.add(), fc1.sub(), fc1.mul(), fc1.div())
print(type(fc2))
```

```python
# 객체의 멤버(변수, 함수(메서드)) 참조
# 변수.변수  , 변수.함수
# (메모리주소값).

fc1.setData(10,20) # 변수 생성과 초기화
print(fc1.first, fc1.second)
```

```python
# 클래스의 상속 : 새로운 클래스를 만드는 것!
# 기존의 클래스의 멤버들을 받아서 새로운 클래스를 생성하는 방법
# class 클래스 이름(부모클래스)
# MoreFourCal

class MoreFourCal(FourCal) :
    #pass

    def pow(self):
        return self.first**self.second
    
    # 오버라이딩: 함수의 이름과 매개변수는 그대로 하고 처리 부분을 수정해서 사용 가
    def div(self) :
        if self.second == 0:
            return 0
        return self.first / self.second
```

```python
cal3 = MoreFourCal(10,0)
```
