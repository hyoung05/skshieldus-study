01 1 부터 100 까지의 합을 구하는 프로그램 작성

```python
total = 0

for a in range(1,101) :
    
    total += a

print(total)
```

02 for문을 이용하여 1부터 10까지 곱한 결과를 출력하는 프로그램 작성

```python
result = 1

for i in range(1, 11) :
    result *= i

print(result)
```

03 구구단의 짝수 단(2,4,6,8)만 출력하는 프로그램 적성

```python
# 단, 2단은 2x2, 4x4 6x6 8x8까지 출력

dan = 2
for dan in range(2, 9):

    if dan % 2 == 0: 
        print(f'{dan}단')

        if dan == 2:
            for i in range(dan+1):
                result = dan * i
                print(f'{dan} x {i} = {result}')

        elif dan == 4:
            for i in range(dan+1):
                result = dan * i
                print(f'{dan} x {i} = {result}')

        elif dan == 6:
            for i in range(dan+1):
                result = dan * i
                print(f'{dan} x {i} = {result}')

        elif dan == 8:
            for i in range(dan+1):
                result = dan * i
                print(f'{dan} x {i} = {result}')

    else:
        print('')

        

        
        
```

04 while문을 무한 루프로 구성하여 작성하는 예제

```python
# 1부터 시작해서 모든 짝수와 3의 배수를 더해서 그 합이 언제 1000이 넘어서는지, 
# 그리고 1000이 넘어선 값은 얼마가 되는지 계산하여 출력

num = 1
result = 0

while True:
    num += 1

    if num % 2 == 0 or num % 3 == 0:
        result += num
        
        if result >= 1000 :
            print(num)
            print(result)

            break

            




```
